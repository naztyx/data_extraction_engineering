```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
chrome_options = webdriver.ChromeOptions()
```

```python
chrome_options.binary_location = 'C:/chromedriver/chrome-win64/chrome.exe'
```

```python
chrome_options.add_argument("--disable-notifications")
```

```python
chrome_options.add_argument('--start-maximized')
```

```python
service = Service('C:/chromedriver/chromedriver-win64/chromedriver.exe')
```

```python
url = 'https://www.eosworldwide.com/implementer/page/'
```

```python
name = []
elementor_title = []
email = []
phone_number = []
address = []
service_area = []
linkedin = []
eos_link = []
```

```python
collected_names = set()
```

```python
collected_names
```

```python
%%time

# initialize the driver
driver = webdriver.Chrome(service=service,options=chrome_options)

page_num = 1

attempt = 0

while True:
    try:    
        driver.get(f'{url}{page_num}')
        driver.implicitly_wait(2)

        next_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/div/section[2]/div/div[2]/div/div[3]/div/span/div/div[30]/div/div/section/div')))
       
        driver.execute_script('arguments[0].scrollIntoView(true)', next_element)
        time.sleep(2)
        
        top_level_collection = driver.find_element(By.CSS_SELECTOR, 'span.geodir-loop-container.bsui.sdel-8815e09d')
        each_implementer = top_level_collection.find_elements(By.CSS_SELECTOR, 'div.card.h-100.p-0.m-0.mw-100')
        
        for ind, each in enumerate(each_implementer):
            details = each.text
            if details != '':            
                details = details.split('\n')
                _name = details[1]
                # print(_name)
                if _name not in collected_names:
                    collected_names.add(_name)
                    title = details[0]
                    area = details[2].replace('Service Area: ', '')
                    addr = details[3].replace('Address: ', '')
                    
                    link = each.find_element(By.TAG_NAME, 'a').get_attribute('href')
    
                    try:
                        driver.get(f'{link}#contact')
                        driver.implicitly_wait(3)
                        
                        email_div = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section[15]/div/div[1]/div/div[5]/div/h5/a').get_attribute('href')
                        _email = email_div.split(':')[1]
                        linkedin_div = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/section[15]/div/div[1]/div/div[8]/div/div/span[1]/a')
                        _linkedin = linkedin_div.get_attribute('href')
                        
                        while True:
                            driver.implicitly_wait(2)                       
                            ph_number_div = driver.find_element(By.CSS_SELECTOR, 'h4.elementor-heading-title.elementor-size-default')
                            _ph_number = ph_number_div.text
                            if _ph_number != '':
                                ph_number = _ph_number
                                break
                            else:
                                driver.refresh()
                            
                        driver.back()

                        name.append(_name)
                        elementor_title.append(title)
                        email.append(_email)
                        phone_number.append(ph_number)
                        address.append(addr)
                        service_area.append(area)
                        linkedin.append(_linkedin)
                        eos_link.append(link)

                        print(f'{ind+1}.\nName: {_name}\nTitle: {title}\nPhone Number: {ph_number}\nEmail: {_email}\nLinkedIn: {_linkedin}\nService Area: {area}\nAddress: {addr}\nEOS Link: {link}')

                    except Exception:
                        print('refreshing the implementor for {name} link...')
                        driver.refresh() 
                else:
                    pass               
      
        page_num +=1
        if page_num > 10:
            break
        time.sleep(1)
    except Exception as e:
        print('\nRetrying...\n')
        attempt += 1
        time.sleep(1)
    # finally:
    #     driver.quit()
driver.quit()
```

## Save file to CSV

```python
implementors_details = {
    'name': name,
    'elementor_title': elementor_title,
    'email': email,
    'phone_number': phone_number,
    'address': address,
    'service_area': service_area,
    'linkedin': linkedin,
    'eos_link': eos_link
}
```

```python
eos_implementers_df = pd.DataFrame(implementors_details)
```
